### FaceRecognition.py Functions

#### `face_confidence(face_distance, threshold=0.6)` 

Converts a raw face distance metric into a human-readable confidence percentage.

**Parameters:**
- `face_distance` (float): Distance from facial encoding comparison (0-1 scale, lower is better)
- `threshold` (float): Confidence threshold for calculations (default: 0.6)

**Returns:** Confidence percentage (0-100)

**Example:**
```python
from backend.FaceRecognition import face_confidence

distance = 0.35
confidence = face_confidence(distance)
print(f"Match confidence: {confidence}%")  
# Output: Match confidence: 92.45%
```

---

#### `load_profiles(api_url="http://127.0.0.1:8000/students")`

Fetches all student profiles from the backend API.

**Parameters:**
- `api_url` (str): URL of the student endpoint (default: local server)

**Returns:** List of student profile dictionaries

**Example:**
```python
from backend.FaceRecognition import load_profiles

profiles = load_profiles()
print(f"Total students: {len(profiles)}")
for profile in profiles:
    print(f"  • {profile['Student_Name']}")
```

---

#### `compute_encodings(profiles)`

Pre-computes facial encodings from student images. **Critical optimization performed once at startup.**

**Why This Matters:** 
- Encodings are computed once and cached in memory
- Dramatically speeds up real-time face matching
- Eliminates re-encoding overhead on every match

**Parameters:**
- `profiles` (list): Student profiles from API

**Returns:** Tuple of (known_encodings list, known_profiles list)

**Example:**
```python
from backend.FaceRecognition import load_profiles, compute_encodings

profiles = load_profiles()
encodings, profile_data = compute_encodings(profiles)
print(f"Generated {len(encodings)} encodings")
```

---

#### `match_face(rgb_frame, known_encodings, known_profiles, threshold=0.47)`

Matches a face from an image against the database of known student faces.

**Parameters:**
- `rgb_frame` (numpy array): Image in RGB format
- `known_encodings` (list): Pre-computed face encodings
- `known_profiles` (list): Corresponding student profiles
- `threshold` (float): Matching sensitivity (lower = stricter, default 0.47)

**Returns:** Tuple of (profile dict or "Unknown", match distance)

**Threshold Reference:**
- `0.30-0.40`: Very strict (high false negatives)
- `0.47`: **Recommended** (optimal balance)
- `0.60+`: Very lenient (high false positives)

**Example:**
```python
import cv2
from backend.FaceRecognition import (
    load_profiles,
    compute_encodings,
    match_face,
    face_confidence
)

# Initialize system
profiles = load_profiles()
encodings, profile_data = compute_encodings(profiles)

# Load and process image
image = cv2.imread("student.jpg")
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Find match
profile, distance = match_face(rgb_image, encodings, profile_data)

# Display results
if profile != "Unknown":
    confidence = face_confidence(distance)
    print(f"Match: {profile['Student_Name']}")
    print(f"Confidence: {confidence}%")
else:
    print("No match found")
```

### Example 2: Complete Face Recognition Workflow

````python
import cv2
from backend.FaceRecognition import (
    load_profiles,
    compute_encodings,
    match_face,
    face_confidence
)

# Step 1: Load all student profiles
print("Step 1: Loading student profiles...")
profiles = load_profiles()
print(f"✓ Successfully loaded {len(profiles)} profiles\n")

# Step 2: Pre-compute face encodings
print("Step 2: Computing face encodings...")
known_encodings, known_profiles = compute_encodings(profiles)
print(f"✓ Generated {len(known_encodings)} encodings\n")

# Step 3: Load and process image
print("Step 3: Loading test image...")
image_path = "path/to/test_image.jpg"
frame = cv2.imread(image_path)
rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
print(f"✓ Image loaded successfully\n")

# Step 4: Perform face matching
print("Step 4: Performing face recognition...")
profile, distance = match_face(rgb_frame, known_encodings, known_profiles)
print(f"✓ Face matching complete\n")

# Step 5: Display results
if profile != "Unknown":
    confidence = face_confidence(distance)
    print("="*50)
    print("MATCH FOUND!")
    print("="*50)
    print(f"Student Name: {profile['Student_Name']}")
    print(f"Student ID: {profile['Student_Id']}")
    print(f"Course: {profile['Course']}")
    print(f"Year Level: {profile['Year_level']}")
    print(f"Match Distance: {distance:.4f}")
    print(f"Confidence: {confidence}%")
    print("="*50)
else:
    print("No matching student found")
    print(f"Best match distance: {distance:.4f}")
````

### Example 3: Understanding Confidence Metrics

````python
from backend.FaceRecognition import face_confidence

# Demonstrate how distances convert to confidence scores
test_distances = [0.2, 0.35, 0.47, 0.6, 0.8]
threshold = 0.47

print("\nFace Distance to Confidence Conversion")
print("="*60)
print(f"Threshold: {threshold} (below = match)\n")
print(f"{'Distance':<12} {'Confidence':<15} {'Result':<20}")
print("-"*60)

for distance in test_distances:
    confidence = face_confidence(distance, threshold)
    result = "✓ MATCH" if distance < threshold else "✗ NO MATCH"
    print(f"{distance:<12.2f} {confidence:<14.2f}% {result:<20}")

print("\n" + "="*60)
print("Key Insights:")
print("• Lower distance = better facial match")
print("• Threshold determines matching sensitivity")
print("• Confidence percentage = match reliability")
print("• Default 0.47 provides optimal balance")
print("="*60)
````

### Example 4: Batch Processing Images

````python
import os
import cv2
from pathlib import Path
from backend.FaceRecognition import (
    load_profiles,
    compute_encodings,
    match_face,
    face_confidence
)

# Initialize system
print("Initializing face recognition system...")
profiles = load_profiles()
known_encodings, known_profiles = compute_encodings(profiles)
print(f"System ready. Processing images...\n")

# Process multiple images
image_folder = "path/to/images"
results = []

if os.path.exists(image_folder):
    image_files = list(Path(image_folder).glob("*.jpg"))
    
    for index, image_file in enumerate(image_files, 1):
        print(f"Processing [{index}/{len(image_files)}]: {image_file.name}")
        
        # Load and convert
        frame = cv2.imread(str(image_file))
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Match
        profile, distance = match_face(rgb_frame, known_encodings, known_profiles)
        
        # Store result
        if profile != "Unknown":
            result = {
                "image": image_file.name,
                "status": "Matched",
                "student_name": profile['Student_Name'],
                "student_id": profile['Student_Id'],
                "confidence": face_confidence(distance)
            }
        else:
            result = {
                "image": image_file.name,
                "status": "Not Found",
                "student_name": None,
                "student_id": None,
                "confidence": 0
            }
        
        results.append(result)
        print(f"  → {result['status']}: {result['student_name'] or 'Unknown'}\n")

    # Summary
    print("\n" + "="*60)
    print("BATCH PROCESSING SUMMARY")
    print("="*60)
    matched = sum(1 for r in results if r['status'] == 'Matched')
    unmatched = len(results) - matched
    print(f"Total images: {len(results)}")
    print(f"Matched: {matched}")
    print(f"Not found: {unmatched}")
    print(f"Success rate: {(matched/len(results)*100):.1f}%")
    print("="*60)
````

### Issue: Students Not Found in Database

**Symptoms:** API returns empty results or "Student not found" errors

**Possible Causes:**
- Database hasn't been initialized yet
- No student data has been loaded into the database
- Database file is corrupted

**Solutions:**

1. **Check if database exists:**
   - Look for `backend/students.db` in your project folder
   - If missing, it will be created automatically when the API starts

2. **Verify database initialization:**
   - When you start the API server, it automatically runs `init_db()`
   - Check the server console for any errors during startup

3. **Check if data was seeded:**
   - If you have `data/studentprofiles.json`, you'll need to migrate this data to the database
   - Use a data migration script or manual database insertion

4. **Query the database directly (Python):**
   ```python
   from backend.database import get_connection

   cursor = conn.cursor()
   ```

   - Ensure good front-facing lighting
   - Keep face centered in frame
   - Avoid angles > 45 degrees

2. **Adjust threshold for more lenient matching:**
   ```python
   profile, distance = match_face(
       rgb_frame,
       encodings,
       profiles,
       threshold=0.50  # Increased from 0.47
   )
   ```

   ```python
   import face_recognition
   ```

### Issue: Database Connection Error

**Error:** `sqlite3.OperationalError` or cannot connect to database

**Possible Causes:**
- Database file is locked by another process
- Permission issues on the database file
- Corrupted database file

**Solutions:**

1. **Verify the database path:**
   ```python
   from backend.database import DB_PATH
   print(f"Database location: {DB_PATH}")
   ```

2. **Close any open connections:**
   - Make sure no other Python processes are using the database
   - Restart your API server to reset connections

3. **Check file permissions:**
   - Ensure you have read/write permissions on the `backend/` directory
   - On Windows: Right-click folder → Properties → Security

4. **Backup and reset the database:**
   ```bash
   # Backup current database
   copy backend\students.db backend\students.db.bak
   
   # Delete the old database (it will be recreated)
   del backend\students.db
   
   # Restart the API server to initialize fresh
   uvicorn backend.api:app --reload --port 8000
   ```
